```python
%matplotlib inline
from pylab import *
from os import path
from matplotlib.colors import LogNorm
from mcstas_reader import McSim
```

# McStas results

```python
figure(figsize=(15,4))

subplot(131)
McSim('mcstas/hexagonal_spheres_05m')['detector'].plot(norm=LogNorm(16e-3, 16e3, clip=True), cmap='gist_ncar', cbar=True)
title('Hexagonal arrangement 5m collimation')

subplot(132)
McSim('mcstas/hexagonal_spheres_10m')['detector'].plot(norm=LogNorm(4e-3, 4e3, clip=True), cmap='gist_ncar', cbar=True)
title('Hexagonal arrangement 10m collimation')

subplot(133)
McSim('mcstas/hexagonal_spheres_20m')['detector'].plot(norm=LogNorm(1e-3, 1e3, clip=True), cmap='gist_ncar', cbar=True)
title('Hexagonal arrangement 20m collimation')

tight_layout()
savefig(path.join('plots', 'hexagonal_spheres_mcstas.png'), dpi=300)
```

```python
figure(figsize=(15,4))

subplot(131)
McSim('mcstas/silica_100nm_air_05m')['detector'].plot(norm=LogNorm(16e-3, 16e3, clip=True), cmap='gist_ncar', cbar=True)
title('Silica spheres 5m collimation')

subplot(132)
McSim('mcstas/silica_100nm_air_10m')['detector'].plot(norm=LogNorm(4e-3, 4e3, clip=True), cmap='gist_ncar', cbar=True)
title('Silica spheres 10m collimation')

subplot(133)
McSim('mcstas/silica_100nm_air_20m')['detector'].plot(norm=LogNorm(1e-3, 1e3, clip=True), cmap='gist_ncar', cbar=True)
title('Silica spheres 20m collimation')

tight_layout();
savefig(path.join('plots', 'silica_100nm_air_mcstas.png'), dpi=300)
```
